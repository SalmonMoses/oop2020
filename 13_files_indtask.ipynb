Якщо один із об'єктів знаходиться на будь-якої вісі, тоді результатом повинно бути **false**.      
    
Приклад:

```python
# стартовий код
import math
# для застосування стандартної функції sqrt 

def sign(n):
    if n > 0:
        return 1
    elif n < 0:
        return -1
    else: 
        return 0

class Point:
    """ клас об'єктів, що визначають крапки на площині
    """
    
    def __init__(self, init_x, init_y):
        """ конструктор для класу об'єктів, який визначає крапку 
            з координатами init_x, init_y)
        """
        self.x = init_x
        self.y = init_y
        
    def distance_from_start(self):
        """ визначає дистанцію крапки 
            (self) від початку координат(0, 0))
        """
        return math.sqrt(self.x * self.x + self.y * self.y)
    
    def move_up(self, amount):
        """ зсуває крапку (self) догори 
            на визначену величину 
        """
        self.y += amount

    def move_down(self, amount):
        """ зсуває крапку (self) вниз 
            на визначену величину 
        """
        self.y -= amount
        
    def quarter(self):
        """визначає квадрант на площині, в якому знаходиться точка self.
        Якщо точка знаходиться на однією з вісей, або у точці (0, 0), то метод повертає 'NO'
        """
        signs = sign(self.x), sign(self.y)
        if signs == (1, 1):
            return 'I'
        elif signs == (-1, 1):
            return 'II'
        elif signs == (-1, -1):
            return 'III'
        elif signs == (1, -1):
            return 'IV'
        else:
            return 'NO'
        
    def change(self):
        self.x = -self.x
        self.y = -self.y
        
    def this_quarter(self, other):
        self_quarter = self.quarter()
        other_quarter = other.quarter()
        if self_quarter == 'NO' or other_quarter == 'NO':
            return False
        return self_quarter == other_quarter
        
    def __eq__(self, other):
        if not isinstance(other, Point):
            return false
        return self.x == other.x and self.y == other.y
    
    def __repr__(self):
        return f'({self.x}, {self.y})'
    
point1 = Point(6, 8)
point2 = Point(6, -8)
point3 = Point(-6, -8)
point4 = Point(-6, 8)
point5 = Point(0, 8)

print(point1.x, point1.y, point2.x, point2.y, sep="\n")
print()

print(point1, point2, sep="\n")
print()

print(point1.distance_from_start(), point2.distance_from_start(), sep="\n")
print()

point1.move_down(5)
print(point1)
print()

print(point1.quarter(), point2.quarter(), point3.quarter(), point4.quarter(), point5.quarter(), sep="\n")
print()

point6 = Point(6, -8)
print(point1 == point6, point2 == point6, sep="\n")
print()

point1.change()
point2.change()
print(point1, point2, sep="\n")
print()

print(point1.this_quarter(point2))
```
